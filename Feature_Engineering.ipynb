1. What is a parameter?

A parameter is a value that a machine learning model learns during training. These values help the model make accurate predictions. In simple terms, parameters define the behavior of the model and are adjusted to best fit the data.

2. What is correlation? What does negative correlation mean?

Correlation is a statistical measure that shows how two variables move in relation to each other. If two things increase or decrease together, they have a positive correlation. If one increases while the other decreases, it’s a negative correlation.
A negative correlation means that as one variable increases, the other decreases. For example, as temperature drops, heating costs usually go up — they move in opposite directions.

3. Define Machine Learning. What are the main components in Machine Learning?

Machine Learning is a field of AI where computers learn patterns from data and make decisions without being explicitly told what to do.

Main components include:

Data: Information the model learns from.

Model: The algorithm that learns patterns.

Training: The process of learning from data.

Loss function: Measures how far predictions are from actual results.

Optimizer: Adjusts the model to reduce errors.

4. How does loss value help in determining whether the model is good or not?

The loss value shows how far off the model's predictions are from the actual results. A lower loss means better performance. If the loss is high, the model is making more mistakes.

 5. What are continuous and categorical variables?

Continuous variables are numbers that can take any value within a range (like height, weight, income).

Categorical variables are labels or groups (like gender, color, or type of product).

 6. How do we handle categorical variables in Machine Learning? What are the common techniques?

Machine Learning models need numbers, so we convert categories into numerical form. Common techniques include:

Label Encoding: Assigns a number to each category.

One-Hot Encoding: Creates separate columns for each category with 0s and 1s.

 7. What do you mean by training and testing a dataset?

Training a dataset means using it to teach the model to understand patterns.

Testing a dataset means checking how well the model performs on new, unseen data. It helps us know if the model will work in real-world situations.

 8. What is sklearn.preprocessing?

It is a part of the Scikit-learn library that provides tools to prepare or clean data before using it in a model. It includes methods for scaling, encoding, and transforming features to make them suitable for machine learning.

 9. What is a Test set?

A test set is a portion of the data that the model hasn’t seen during training. It’s used to evaluate how well the model performs on new data.

We divide the data into two parts:

A training set to teach the model.

11. Why do we have to perform EDA before fitting a model to the data?

EDA (Exploratory Data Analysis) helps us understand the data before modeling. It reveals patterns, missing values, and outliers. This step ensures we feed good quality, meaningful data into the model.

Negative correlation means that as one variable goes up, the other goes down. For example, as speed increases, time taken decreases.

14. How can you find correlation between variables in Python?

 By using tools to calculate correlation coefficients, such as Pearson’s, you can measure how strongly two variables are related.

 15. What is causation? Explain difference between correlation and causation with an example.

Causation means one event directly causes another.

Correlation just means two things move together, but one doesn't necessarily cause the other.

Example:
Ice cream sales and sunburns increase together — they’re correlated. But eating ice cream doesn’t cause sunburns. The real cause is hot weather.

 16. What is an Optimizer? What are different types of optimizers? Explain each with an example.

An optimizer helps improve the model by adjusting its parameters to reduce errors.

Common types:

SGD (Stochastic Gradient Descent): Updates the model slowly using one data point at a time.

Adam: Adjusts learning rate automatically; fast and widely used.

RMSProp: Adapts learning based on recent changes; works well for unstable problems.

Each tries to find the best model with the lowest possible error.

 21. What is feature scaling? How does it help in Machine Learning?

Feature scaling is the process of adjusting the range of values of numerical features so they are on a similar scale. It helps models learn better and faster, especially those sensitive to differences in scale (like SVM, KNN).

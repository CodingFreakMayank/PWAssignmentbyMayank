Question 1: What is a random variable in probability theory?

Answer:
A random variable is just a way to give numbers to outcomes of a random event.
Example: Toss a coin → if heads = 1, tails = 0. Here, the random variable is the number we assign (0 or 1).


---

Question 2: What are the types of random variables?

Answer:
There are two main types:

1. Discrete random variable → Can take only certain separate values (like 0,1,2,3…). Example: Number of chocolates in a box.


2. Continuous random variable → Can take any value in a range. Example: Your exact height (like 173.2 cm, 173.25 cm, etc.).




---

Question 3: Difference between discrete and continuous distributions.

Answer:

Discrete distribution: Deals with things we can count. Probabilities are given to each value. Example: Rolling a die (1,2,3,4,5,6).

Continuous distribution: Deals with things we measure. Probabilities are spread across ranges, not single points. Example: Time taken to run 100m.



Question 4: What is a binomial distribution, and how is it used in probability?

Answer:
A binomial distribution is a probability distribution that shows the chance of getting a certain number of successes in a fixed number of trials, where each trial has only two outcomes (success or failure).
Example: Tossing a coin 10 times and finding the probability of getting exactly 6 heads.


---

Question 5: What is the standard normal distribution, and why is it important?

Answer:
The standard normal distribution is a bell-shaped curve with:

Mean = 0

Standard deviation = 1
It is important because many real-life data follow a normal pattern, and using the standard normal makes it easy to calculate probabilities with Z-scores.



---

Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answer:
The Central Limit Theorem says that if we take many samples from any population, the average of those samples will follow a normal distribution (bell curve), no matter how the original population looks — provided the sample size is large enough.


Question 7: What is the significance of confidence intervals in statistical analysis?


Answer:
A confidence interval gives a range of values where we expect the true population value (like the real mean) to lie, with a certain level of confidence (like 95%).

Example: If we say the average height is 173 ± 2 cm with 95% confidence, it means we are 95% sure the true average is between 171 and 175 cm.


---


Question 8: What is the concept of expected value in a probability distribution?


Answer:
The expected value is the long-term average outcome if we repeat an experiment many times.

Example: If you roll a fair die, the expected value is (1+2+3+4+5+6)/6 = 3.5. It doesn’t mean you’ll roll 3.5, but on average, that’s what the outcome trends towards.



Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.

Solution -

```python

'''Question 9: Write a Python program to generate 1000 random numbers from a normal

distribution with mean = 50 and standard deviation = 5. Compute its mean and standard

deviation using NumPy, and draw a histogram to visualize the distribution.'''


import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
mean = 50
std_dev = 5
data = np.random.normal(mean, std_dev, 1000)

# Compute mean and standard deviation
calculated_mean = np.mean(data)
calculated_std = np.std(data)

print("Calculated Mean:", calculated_mean)
print("Calculated Standard Deviation:", calculated_std)

# Draw histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title("Normal Distribution (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python

'''  10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.'''




import numpy as np
import scipy.stats as stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
data = np.array(daily_sales)

# Sample size, mean, and standard deviation
n = len(data)
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)  # sample std (ddof=1)

# Standard error
se = std_sales / np.sqrt(n)

# 95% confidence interval (using Z = 1.96)
confidence = 0.95
z_score = stats.norm.ppf((1 + confidence) / 2)
margin = z_score * se
ci_lower = mean_sales - margin
ci_upper = mean_sales + margin

print("Mean Sales:", mean_sales)
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(ci_lower, ci_upper))
```

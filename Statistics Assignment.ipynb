Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Answer: Descriptive statistics summarize and describe the main features of a dataset. They provide simple summaries about the sample and the measures.

Example: If a class of 30 students took a test and the average score was 75, this is descriptive statistics — it summarizes the scores in that specific group.

Inferential statistics use a sample of data to make predictions or generalizations about a larger population.


Example: If you survey 100 people about their favorite ice cream and conclude that of all people in your city prefer chocolate.


Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Answer: Sampling is the process of selecting a subset (sample) from a larger group (population) to draw conclusions about the whole population.
Random sampling is where every individual in the population has an equal chance of being selected. It’s simple and unbiased but may not represent all subgroups equally.


Stratified sampling involves dividing the population into distinct subgroups (strata) based on a characteristic (e.g., age, gender), then randomly sampling from each subgroup. This ensures representation from all important groups.


Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

Answer:
Mean is the average of all values. It  adds all values and divides them by the number of values.


Median is the middle value when data is arranged in order.
Mode is the value that appears most frequently in a dataset.
These measures are important because they help summarize large datasets with a single value, giving insight into the central point or typical value of the data.


Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

Answer: Skewness measures the asymmetry of the data distribution.
Positive skew (right-skewed): Tail is longer on the right side. Most values are on the left.
Example: Income distribution, where most people earn less, and few earn much more.
Negative skew (left-skewed): Tail is longer on the left side.
Kurtosis measures the "tailedness" or peakedness of the distribution.
High kurtosis (leptokurtic): Sharp peak and fat tails.
Low kurtosis (platykurtic): Flat peak and thin tails.
A positive skew implies that the data has a longer right tail, meaning there are more lower values and a few very high values that pull the mean to the right of the median.


```python
''' Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.
 numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
(Include your Python code and output in the code box below.) '''

#solution;

import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean = statistics.mean(numbers)
median = statistics.median(numbers)
mode = statistics.mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
'''Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]'''


import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Covariance
covariance_matrix = np.cov(list_x, list_y, bias=False)
covariance = covariance_matrix[0][1]

# Correlation
correlation_matrix = np.corrcoef(list_x, list_y)
correlation = correlation_matrix[0][1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)



```

```python
'''Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
● Explain how you would use covariance and correlation to explore this relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

(Include your Python code and output in the code box below.)
'''

import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Covariance
cov_matrix = np.cov(advertising_spend, daily_sales, bias=False)
cov = cov_matrix[0][1]

# Correlation
corr_matrix = np.corrcoef(advertising_spend, daily_sales)
corr = corr_matrix[0][1]

print("Covariance:", cov)
print("Correlation Coefficient:", corr)




'''Interpretation:

Interpretation:
Strong positive correlation — higher advertising spend is associated with higher sales.'''
```

```python
'''Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7] '''

import matplotlib.pyplot as plt
import statistics

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Summary statistics
mean = statistics.mean(survey_scores)
std_dev = statistics.stdev(survey_scores)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Histogram
plt.hist(survey_scores, bins=6, edgecolor='black')
plt.title("Histogram of Survey Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()





'''Interpretation:
The scores are concentrated around 7–9, indicating generally high satisfaction with some spread'''
```
